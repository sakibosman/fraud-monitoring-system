```python
import pandas as pd
```

```python
# Dataset: Canadian Anti-Fraud Centre Reporting Data
df = pd.read_csv("C:/Users/User/Desktop/fraud-monitoring-system/data/fraud_data.csv")
```

```python
print("Preview:")
print(df.head())

print("\nInformation:")
print(df.info())
```

```python
# Overview
print("\nTotal Reports:", len(df))
print("Total Dollar Loss:", df["dollar_loss"].sum())
print("Average Loss:", df["dollar_loss"].mean())
```

```python
# Fraud by Categories
fraud_by_category = (
    df.groupby("fraud_category")["dollar_loss"]
    .sum()
    .sort_values(ascending=False)
)

print("\nFraud Loss by Category:")
print(fraud_by_category)
```

```python
# Fraud by Province
fraud_by_province = (
    df.groupby("province_state")["dollar_loss"]
    .sum()
    .sort_values(ascending=False)
)

print("\nFraud Loss by Province:")
print(fraud_by_province)
```

```python
# Fraud Method vs Dollar Loss
fraud_by_method = (
    df.groupby("method")["dollar_loss"]
    .sum()
    .sort_values(ascending=False)
)

print("\nFraud Loss by Method:")
print(fraud_by_method)
```

```python
# Fraud by Age Range

fraud_by_age = (
    df.groupby("age_range")["dollar_loss"]
    .sum()
    .sort_values(ascending=False)
)

print("\nFraud Loss by Age Range:")
print(fraud_by_age)
```

```python
# Monthly Fraud Trends
df["date_received"] = pd.to_datetime(df["date_received"])
monthly_trend = (
    df.groupby(df["date_received"].dt.to_period("M"))["dollar_loss"]
    .sum()
)

print("\nMonthly Fraud Trend:")
print(monthly_trend)
```

```python
# Top 10 Largest Fraud Cases
top_cases = df.sort_values("dollar_loss", ascending=False).head(10)

print("\nTop 10 Largest Fraud Cases:")
print(top_cases)
```

```python
# Cases With Financial Loss (%)
cases_with_loss = (df["dollar_loss"] > 0).sum()
total_cases = len(df)
percentage_loss = (cases_with_loss / total_cases) * 100

print("\nCases with Financial Loss:", cases_with_loss)
print("Percentage of Cases with Loss:", round(percentage_loss, 2), "%")
```
